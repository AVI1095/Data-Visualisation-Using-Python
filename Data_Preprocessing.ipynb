```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
data = {
    "Student_ID": [101, 102, 103, 104, 105, 105, 106],
    "Name": ["Aarav", "isha", "Rohan", "Meera", None, "Kabir", "   Aarav"],
    "Age": [20, None, 19, "22", 20, 20, 19],
    "Department": ["CSE", "it", "ECE", "CSE", "mech", "ME", "CSE "],
    "Marks": [85, 90, None, 88, 76, 76, 101],
    "Attendance(%)": [92, "85%", 80, None, 70, 70, 110]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.info())
```

```python
print(df.describe(include="all"))
```

```python
print(df.isnull().sum())
```

```python
df["Name"] = df["Name"].astype(str).str.strip()
df["Department"] = df["Department"].astype(str).str.strip()
```

```python
df
```

```python
df["Name"] = df["Name"].fillna("Unknown")
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["Attendance(%)"] = df["Attendance(%)"].astype(str).str.replace("%", "", regex=False)
df["Attendance(%)"] = pd.to_numeric(df["Attendance(%)"], errors="coerce")
df["Attendance(%)"] = df["Attendance(%)"].fillna(df["Attendance(%)"].median())
```

```python
df = df.drop_duplicates(subset=["Student_ID", "Name"], keep="first")
```

```python
df["Marks"] = df["Marks"].clip(0, 100)
df["Attendance(%)"] = df["Attendance(%)"].clip(0, 100)
```

```python
scaler = StandardScaler()
df["Marks_scaled"] = scaler.fit_transform(df[["Marks"]])
print(df[["Marks","Marks_scaled"]])
```
