```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r"/Users/adityaingle/Documents/csv files ML/Social_Network_Ads.csv")
```

```python
print("First 5 Rows:\n", data.head())
```

```python
print("\nDataset Info:\n")
print(data.info())
```

```python
print("\nSummary Statistics:\n", data.describe())
```

```python
print("\nGender Count:\n", data['Gender'].value_counts())
```

```python
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
print("\nCorrelation Matrix:\n", data.corr())
```

```python
gender_count = data['Gender'].value_counts()
plt.bar(gender_count.index, gender_count.values, color=['blue','orange'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(data['Age'], bins=15, color='green', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.plot(data['EstimatedSalary'][:50], marker='o', linestyle='--', color='purple')
plt.title("Estimated Salary (First 50 Users)")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.show()
```

```python
purchase_count = data['Purchased'].value_counts()
plt.pie(purchase_count, labels=['Not Purchased','Purchased'], autopct='%1.1f%%', colors=['red','yellow'])
plt.title("Purchase Distribution")
plt.show()
```

```python
plt.scatter(data['Age'], data['EstimatedSalary'], c=data['Purchased'], cmap='coolwarm')
plt.title("Age vs Estimated Salary (colored by Purchase)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```
